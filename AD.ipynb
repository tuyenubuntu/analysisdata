```python
import os
import pandas as pd
import openpyxl
from openpyxl import load_workbook, Workbook
import warnings

def load_data(path: str, sheet_name):
    """Load data from an Excel file."""
    warnings.simplefilter(action='ignore', category=UserWarning)
    return pd.read_excel(path, sheet_name)
path_twi = 'TWI_2023_1.xlsx'
path_overal = 'Overall_2023.xlsx'

'''
df_twi = pd.read_excel(path_twi, sheet_name = 3)
df_overal = pd.read_excel(path_overal, sheet_name = 3)
'''

df_twi = load_data(path_twi, sheet_name = 3)
df_overal = load_data(path_overal, sheet_name = 4)

wb_twi = load_workbook(path_twi)
wb_overal = load_workbook(path_overal)
```

```python
ws_twi = wb_twi['Phong Export']
ws_overal = wb_overal['TWI']
```

```python
ls_twi = list(df_twi['Trigger process number/ Số TWI:'])
ls_twi_start = list(df_twi['Start'])
ls_twi_end = list(df_twi['End'])
ls_twi_duration = list(df_twi['Duration'])
```

```python
ls_overal = list(df_overal['Trigger process number/ Số TWI:'])
```

```python
fix_code = set(ls_twi) & set(ls_overal)
len(list(fix_code))
```

```python

for i in fix_code:
    ind_twi = ls_twi.index(i)
    ind_overal = ls_overal.index(i)+2
    ws_overal[f'I{ind_overal}'] = ls_twi_start[ind_twi]
    ws_overal[f'J{ind_overal}'] = ls_twi_end[ind_twi]
    ws_overal[f'K{ind_overal}'] = ls_twi_duration[ind_twi]
wb_overal.save (path_overal)
print ('Done!')
```
