```python
import os
import pandas as pd
import openpyxl
from openpyxl import load_workbook, Workbook
from openpyxl.styles import PatternFill
from datetime import datetime
import warnings

#path for 2 excel file analysis
path_twi = 'TWI_2023_1.xlsx'
path_overal = 'Overall_2023.xlsx'

#function for ignore Warning read data in file excel
def load_data(path: str, sheet_name):
    """Load data from an Excel file."""
    warnings.simplefilter(action='ignore', category=UserWarning)
    return pd.read_excel(path, sheet_name)

#replace for above function
'''
df_twi = pd.read_excel(path_twi, sheet_name = 3)
df_overal = pd.read_excel(path_overal, sheet_name = 3)
'''
#load data farame 2 excel file (lib of pandas)
df_twi = load_data(path_twi, sheet_name = 3)
df_overal = load_data(path_overal, sheet_name = 4)

#load workbook 2 excel file (lib of openpyxl)
wb_twi = load_workbook(path_twi)
wb_overal = load_workbook(path_overal)
```

```python
# load sheet from 2 excel file (lib of openpyxl)
ws_twi = wb_twi['Phong Export'] #data
ws_overal = wb_overal['TWI'] #data
```

```python
#load data from columns (Trigger process number/ Số TWI:, Start, End, Duration) in excel TWI  (TWI_2023_1.xlsx) , after that convert to type data is list
ls_twi = list(df_twi['Trigger process number/ Số TWI:'])
ls_twi_start = list(df_twi['Start'])
ls_twi_end = list(df_twi['End'])
ls_twi_duration = list(df_twi['Duration'])
```

```python
#load data from columns (Trigger process number/ Số TWI:, Start, End, Duration) in excel overall  (Overall_2023.xlsx)
ls_overal = list(df_overal['Trigger process number/ Số TWI:'])
```

```python
#create list for all code twi to coincide in variable : fix_code
fix_code = set(ls_twi) & set(ls_overal)
len(list(fix_code))
```

```python
#co-ordinate and modify data 
for i in fix_code:
    #return a number of code twi (file excel twi), follow list in pandas function 
    ind_twi = ls_twi.index(i)
    #return a number of code twi (file excel overall), follow co-ordinate in openpyxl function
    ind_overal = ls_overal.index(i)+2
    #all letter I,J,K are columns as Start, End, Duration
    #Co-ordinate (follow row of code) =  content (follow row of code)
    ws_overal[f'I{ind_overal}'] = ls_twi_start[ind_twi]
    ws_overal[f'J{ind_overal}'] = ls_twi_end[ind_twi]
    ws_overal[f'K{ind_overal}'] = ls_twi_duration[ind_twi]
#save file
wb_overal.save (path_overal)
print ('Done!')
```
